```python
import os
from dotenv import load_dotenv
load_dotenv()
# Check if variables are loaded 
groq_key = os.getenv('GROQ_API_KEY')
if not groq_key :
    raise ValueError("Missing required environment variables in .env file")
os.environ['GROQ_API_KEY'] = groq_key
```

```python
from langchain_groq import ChatGroq
LLM=ChatGroq(model='llama-3.1-8b-instant',temperature=0.7, max_tokens=512)
```

```python
from langchain_core.messages import HumanMessage,SystemMessage
messages=[
    SystemMessage(content="Translate the following from English to Hindi."),
    HumanMessage(content="Hello, how are you?")
] 
result=LLM.invoke(messages)
```

```python
result.content
```

```python
from langchain_core.output_parsers import StrOutputParser
parser=StrOutputParser()
parser.invoke(result)
```

```python
#Langchain Expression Language (LCEL) Example
chain=LLM|parser
chain.invoke(messages)
```

```python
from langchain_core.prompts import ChatPromptTemplate

generic_prompt="Translate following into {language}"
prompt=ChatPromptTemplate.from_messages(
    [("system",generic_prompt),("user","{text}")]
    )
```

```python
result=prompt.invoke({"language":"French","text":"Good morning"})
```

```python
result.messages
```

```python
chain=prompt|LLM|parser
chain.invoke({"language":"Spanish","text":"Good night"})
```
