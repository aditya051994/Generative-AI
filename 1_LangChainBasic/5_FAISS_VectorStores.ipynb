```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_text_splitters import CharacterTextSplitter
loader=TextLoader('../Resources/speech.txt')
documents=loader.load()
```

```python
text_splitter=CharacterTextSplitter(chunk_size=1000,chunk_overlap=0)
docs=text_splitter.split_documents(documents)
docs
```

```python
db=FAISS.from_documents(docs, HuggingFaceEmbeddings())
db
```

```python
##query vector database
query="What does speaker belives is the main reason the United State enter the war?"
docs=db.similarity_search(query)
docs
```

```python
db.similarity_search_with_score(query)
```

```python
## Retrival
retriever=db.as_retriever()
docs=retriever.invoke(query)
docs[0].page_content
```

```python
db.save_local("faiss_index")
```

```python
new_db=FAISS.load_local("faiss_index",HuggingFaceEmbeddings(),allow_dangerous_deserialization=True)  
docs=new_db.similarity_search(query) 
docs
```
