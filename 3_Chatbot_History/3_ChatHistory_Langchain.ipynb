```python
import os
from dotenv import load_dotenv
from langchain_core.messages import SystemMessage,trim_messages        ,HumanMessage,AIMessage     
from langchain_groq import ChatGroq
load_dotenv()
# Check if variables are loaded 

groq_key = os.getenv('GROQ_API_KEY')
if not groq_key :
    raise ValueError("Missing required environment variables in .env file")
os.environ['GROQ_API_KEY'] = groq_key

LLM=ChatGroq(model='llama-3.1-8b-instant',temperature=0.7)
trim=trim_messages(
    max_tokens=30,strategy='last',token_counter=LLM,include_system=True,allow_partial=False,start_on="human"        
)
```

```python
messages=[
    SystemMessage(content="you'r good assistant"),
    HumanMessage(content="Hello, I'm bob?"),
    AIMessage(content="Hello Bob! How can I assist you today?"),
    HumanMessage(content="I like Vanila icecream") ,
    AIMessage(content="nice"),
    HumanMessage(content="what is 2+2 ?"),
    AIMessage(content="4") ,
    HumanMessage(content="thanks"),
    AIMessage(content="your welcome"),
    HumanMessage(content="having fun ?"),
    AIMessage(content="yes!")
]
trim.invoke(messages)
```
