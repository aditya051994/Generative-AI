```python
import os
from dotenv import load_dotenv
load_dotenv()
# Check if variables are loaded 
groq_key = os.getenv('GROQ_API_KEY')
if not groq_key :
    raise ValueError("Missing required environment variables in .env file")
os.environ['GROQ_API_KEY'] = groq_key
```

```python
from langchain_groq import ChatGroq
LLM=ChatGroq(model='llama-3.1-8b-instant',temperature=0.7)
```

```python
from langchain_core.messages import HumanMessage, SystemMessage,AIMessage
LLM.invoke([
    SystemMessage(content="My name is Aditya Bhonde and I'm Generative and Agentic Ai developer.")])
```

```python
LLM.invoke([
    HumanMessage(content="My name is Aditya Bhonde and I'm Generative and Agentic Ai developer."),
    AIMessage(content="Nice to meet you, Aditya Bhonde! As a Generative and Agentic AI developer, you're likely working on cutting-edge projects that involve creating AI systems that can learn, adapt, and exhibit intelligent behavior. That's fascinating!\n\nCould you tell me more about your work and what excites you about developing Generative and Agentic AI models? Are you working on any specific projects or applications, such as chatbots, robotics, or autonomous systems?\n\nAlso, I'm happy to chat and help with any questions or topics you'd like to discuss. What's on your mind, Aditya?"),
    HumanMessage(content="Hey tell me what is my name and what do i do?")])
```

## Message History

```python
##Message History
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
```

```python
store={}
def get_session_history(session_id:str)->BaseChatMessageHistory:
    if session_id not in store:
        store[session_id]=ChatMessageHistory()
    return store[session_id]    

with_history=RunnableWithMessageHistory(
    LLM,get_session_history)     
```

```python
config1={"configurable":{"session_id":"user_1"}}
config1
```

```python
with_history.invoke([HumanMessage(content="My name is Aditya Bhonde and I'm Generative and Agentic Ai developer.")], config=config1)
```

```python
with_history.invoke(
    [HumanMessage(content="Hello, can you tell me what is my name and what do I do?")],
    config=config1)
```
